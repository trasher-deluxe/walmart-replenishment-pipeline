```python
import sys

sys.path.append('..')

import numpy as np
import pandas as pd
```

## calendar

```python
df_calendar = pd.read_csv('../data/calendar.csv')
```

```python
print("=" * 50)
print("1. DIMENSIONES DEL DATASET (Filas, Columnas)")
print("=" * 50)
print(df_calendar.shape)

print("\n" + "=" * 50)
print("2. ESTRUCTURA, TIPOS DE DATOS Y USO DE MEMORIA")
print("=" * 50)
df_calendar.info(memory_usage="deep")

print("\n" + "=" * 50)
print("3. PRIMERAS 5 FILAS (HEAD)")
print("=" * 50)
display(df_calendar.head())

print("\n" + "=" * 50)
print("4. ÚLTIMAS 5 FILAS (TAIL)")
print("=" * 50)
display(df_calendar.tail())

print("\n" + "=" * 50)
print("5. DESCRIPCIÓN ESTADÍSTICA GENERAL (VARIABLES NUMÉRICAS)")
print("=" * 50)
display(df_calendar.describe().T)

print("\n" + "=" * 50)
print("6. DESCRIPCIÓN GENERAL (VARIABLES NO NUMÉRICAS)")
print("=" * 50)
try:
    display(df_calendar.describe(include=["object", "category", "bool"]).T)
except ValueError:
    print("No se encontraron variables categóricas u objeto en el DataFrame.")
```

```python
print("=" * 60)
print("1. VALORES NULOS Y PORCENTAJES")
print("=" * 60)
null_df = pd.DataFrame({
    'Total_Nulos': df_calendar.isnull().sum(),
    'Porcentaje_Nulos (%)': (df_calendar.isnull().sum() / len(df_calendar)) * 100
})
display(null_df)

print("\n" + "=" * 60)
print("2. FILAS DUPLICADAS")
print("=" * 60)
print(f"Total de filas duplicadas exactas: {df_calendar.duplicated().sum()}")

print("\n" + "=" * 60)
print("3. CARDINALIDAD Y TIPOS DE DATOS")
print("=" * 60)
card_df = pd.DataFrame({
    'Tipo_Dato': df_calendar.dtypes,
    'Valores_Unicos': df_calendar.nunique(),
    'Cardinalidad (%)': (df_calendar.nunique() / len(df_calendar)) * 100
})
display(card_df)

print("\n" + "=" * 60)
print("4. DETECCIÓN DE ESPACIOS EN BLANCO Y STRINGS EXTRAÑOS")
print("=" * 60)
str_cols = df_calendar.select_dtypes(include=['object', 'string']).columns
for col in str_cols:
    non_null_mask = df_calendar[col].notnull()
    # Verificar espacios al inicio o final
    whitespaces = df_calendar[non_null_mask][col].astype(str).str.strip().ne(df_calendar[non_null_mask][col].astype(str)).sum()
    # Verificar strings vacíos o con solo espacios
    empty_strings = df_calendar[non_null_mask][col].astype(str).str.strip().eq("").sum()
    print(f"Columna '{col}': {whitespaces} con espacios extra | {empty_strings} cadenas vacías")

print("\n" + "=" * 60)
print("5. FRECUENCIA DE CATEGORÍAS (CATEGÓRICAS Y BOOLEANAS)")
print("=" * 60)
cat_cols = df_calendar.select_dtypes(include=['object', 'category', 'bool']).columns
for col in cat_cols:
    if col != 'date':
        print(f"\n--- Frecuencias para '{col}' ---")
        print(df_calendar[col].value_counts(dropna=False))
```

```python
print("=" * 60)
print("1. ESTADÍSTICA DESCRIPTIVA: VARIABLES NUMÉRICAS")
print("=" * 60)

num_cols = df_calendar.select_dtypes(include=['int64', 'float64']).columns.tolist()

num_stats = []
for col in num_cols:
    s = df_calendar[col]
    p25 = s.quantile(0.25)
    p75 = s.quantile(0.75)
    
    num_stats.append({
        'Variable': col,
        'Media': round(s.mean(), 2),
        'Mediana': round(s.median(), 2),
        'Desv_Std': round(s.std(), 2),
        'Min': s.min(),
        'P25': p25,
        'P75': p75,
        'Max': s.max(),
        'Rango': s.max() - s.min(),
        'IQR': round(p75 - p25, 2),
        'Asimetría': round(s.skew(), 2),
        'Curtosis': round(s.kurt(), 2)
    })

num_summary_df = pd.DataFrame(num_stats)
display(num_summary_df)

print("\n" + "=" * 60)
print("2. ESTADÍSTICA DESCRIPTIVA: VARIABLES CATEGÓRICAS Y BOOLEANAS")
print("=" * 60)

cat_cols = df_calendar.select_dtypes(include=['object', 'category', 'bool']).columns.tolist()

for col in cat_cols:
    if col != 'date':
        freq = df_calendar[col].value_counts(dropna=False)
        pct = df_calendar[col].value_counts(dropna=False, normalize=True) * 100
        
        cat_df = pd.DataFrame({
            'Frecuencia_Absoluta': freq,
            'Porcentaje (%)': pct.round(2)
        })
        
        print(f"\n--- Distribución de '{col}' ---")
        display(cat_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración global de estilo
sns.set_theme(style="whitegrid")
plt.rcParams.update({'font.size': 10})

# 1. Distribución de Variables Numéricas (Histogramas + KDE)
num_cols = ['day_of_week', 'week_of_year', 'month', 'year', 'quarter']
fig, axes = plt.subplots(2, 3, figsize=(15, 8))
axes = axes.flatten()

for i, col in enumerate(num_cols):
    sns.histplot(df_calendar[col], kde=True, ax=axes[i], bins=15, color='steelblue', edgecolor='black')
    axes[i].set_title(f'Distribución de {col}', fontsize=12, fontweight='bold')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Frecuencia')

axes[-1].axis('off')  # Ocultar panel sobrante
plt.tight_layout()
plt.show()

# 2. Distribución de Variables Categóricas y Booleanas (Countplots)
cat_cols = ['day_name', 'season', 'is_holiday', 'is_payday', 'is_weekend', 'is_navidad_season']
fig, axes = plt.subplots(2, 3, figsize=(16, 9))
axes = axes.flatten()

for i, col in enumerate(cat_cols):
    sns.countplot(data=df_calendar, x=col, ax=axes[i], palette='Blues_r', order=df_calendar[col].value_counts().index)
    axes[i].set_title(f'Frecuencia de {col}', fontsize=12, fontweight='bold')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Conteo')
    axes[i].tick_params(axis='x', rotation=30 if col in ['day_name', 'season'] else 0)

plt.tight_layout()
plt.show()

# 3. Diagramas de Caja (Boxplots) para Numéricas
fig, axes = plt.subplots(1, 4, figsize=(16, 4))
num_box_cols = ['day_of_week', 'week_of_year', 'month', 'quarter']

for i, col in enumerate(num_box_cols):
    sns.boxplot(y=df_calendar[col], ax=axes[i], color='lightblue')
    axes[i].set_title(f'Boxplot de {col}', fontsize=11, fontweight='bold')

plt.tight_layout()
plt.show()
```
